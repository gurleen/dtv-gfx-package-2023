```python
import os
import pandas as pd
import difflib
```

```python
directory = '../media/logos-white'
names = os.listdir(directory)
names.sort()
names
```

```python
def best_match(x):
    matches = difflib.get_close_matches(x.upper(), names, n=1)
    if len(matches) > 0:
        return matches[0]
    else:
        return None
```

```python
df = pd.read_csv("espn_mbb_teams.csv")
df["logo_name"] = df["short_name"].apply(best_match)
```

```python
df.to_csv("espn_mbb_teams_with_logos.csv", index=False)
```
